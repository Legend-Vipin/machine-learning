Naive Bays Algorithm

```python
import pandas as pd
```

```python
df = pd.read_csv('Pearl_Jam_Tour2.csv', encoding='unicode_escape')
```

```python
df
```

```python
df2 = pd.get_dummies(df[['Continent','Day of Week']])	
```

```python
df2= pd.get_dummies(df, columns=['Continent', 'Day of Week'])
df2 = df2.astype(int)
```

```python
df2
```

```python
df3 = pd.concat([df,df2],axis=1)
```

```python
df3
```

```python
df4 = df3.drop(['Continent','Day of Week'],axis=1)
```

```python
df4
```

```python
x = df4.drop(['Sold Out'],axis=1)
```

```python
y = df4['Sold Out']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred = gnb.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
