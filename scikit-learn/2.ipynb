Features of scaliing standardizatin and normalizatino in scikit-learn

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Read the data
df = pd.read_csv('500hits.csv', encoding='latin-1')

# Create feature matrix, dropping PLAYER and HOF columns and any other non-numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
X = df[numeric_columns]
print(X)

# Select first 13 numeric columns
X1 = X.iloc[:, :13]
X2 = X.iloc[:, :13]

# Initialize scaler
scaleStandard = StandardScaler()
```

```python
df.describe().round(3)
```

```python
X1 = scaleStandard.fit_transform(X1)
```
