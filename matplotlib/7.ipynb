# Box Plot & whisker plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
# plt.boxplot(x,vert=False)
# plt.boxplot(x,widths=0.8)
# plt.boxplot(x,tick_labels=["python"],patch_artist=True,showmeans=True)
# plt.boxplot(x,tick_labels=["python"],showmeans=True,whis=2)
plt.boxplot(x,tick_labels=["python"],sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"))
plt.title("Boxplot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
x1=[10,20,40,50,60,70,90]
y=[x,x1]
plt.boxplot(x,sym="g+",showmeans=True,boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y")
)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [11, 21, 32, 41, 51, 61, 71]
x1 = [10, 20, 40, 50, 60, 70, 90]
y = [x, x1]

plt.boxplot(y, 
            tick_labels=["python", "c++"], 
            sym="g+", 
            showmeans=True,
            boxprops=dict(color="r"),
            capprops=dict(color="r"),
            whiskerprops=dict(color="r"),
            flierprops=dict(markeredgecolor="y"))

plt.show()
```
