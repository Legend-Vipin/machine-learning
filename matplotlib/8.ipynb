# Stack plot

```python
import matplotlib.pyplot as plt
x=[1, 2, 3, 4, 5]
area1=[2,3,2,5,5]
area2 =[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,colors=['r','g','m'])
plt.legend(l,loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1, 2, 3, 4, 5]
area1=[2,3,2,5,5]
area2 =[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,colors=['r','g','m'],baseline="sym")
plt.legend(l,loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1, 2, 3, 4, 5]
area1=[2,3,2,5,5]
area2 =[2,3,4,5,6]
area3=[1,3,2,4,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,colors=['r','g','m'],baseline="wiggle")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(l,loc='upper left')
plt.show()
```
