# Bar Plot  

```python
import matplotlib.pyplot as plt 
x=["python","C","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=18)
plt.ylabel("NO.")
plt.title("Famous languages")
c = ["y","b","m","g"]
# plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5)
plt.bar(x,y,width=0.4,color="r",edgecolor="r",label="wscube")
plt.bar(x,z,width=0.4,color="g",edgecolor="r",label="popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x=["python","C","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
p=np.arange(len(x))
width=0.2
p1=[j+width for j in p]
plt.xlabel("language",fontsize=18)
plt.xlabel("language",fontsize=18)
plt.ylabel("NO.")
plt.title("Famous languages")
c = ["y","b","m","g"]
# plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5)
plt.bar(p,y,width,color="r",edgecolor="r",label="wscube")
plt.bar(p1,z,width,color="g",edgecolor="r",label="popularity")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=["python","C","c++","java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=18)
plt.ylabel("NO.")
plt.title("Famous languages")
c = ["y","b","m","g"]
plt.barh(x,y,color="r",label="wscube")
plt.barh(x,z,color="y",label="popularity")
plt.legend()
plt.show()
```
