# Stem plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt=":")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="ro")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=2,basefmt="g",label="python")
plt.legend()
plt.show()
```
