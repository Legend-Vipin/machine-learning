# Scatter Plot

```python
import matplotlib.pyplot as plt 
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
sizes = [10, 2, 4, 2, 10, 7, 8]
plt.scatter(day, no, color="r", s=sizes)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt 
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,2,6,2]
# colors=["r","y","g","b","r","g","r"]
colors=[10,49,30,29,56,20,30]
sizes = [400, 200, 400, 300, 200, 100, 600]
# plt.scatter(day, no, c=colors, s=sizes,marker="*",edgecolors="g",linewidths=2)
plt.scatter(day, no, c=colors, s=sizes,cmap="viridis",alpha=0.5)
plt.scatter(day, no2, color="r", s=sizes,alpha=0.5)
t=plt.colorbar()
t.set_label("Colorbar")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("No.",fontsize=20)
plt.show()
```
